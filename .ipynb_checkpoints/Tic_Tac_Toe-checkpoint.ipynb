```python
import matplotlib.pyplot as plt
#Clear output
from IPython.display import clear_output
available_boxes = [1,2,3,4,5,6,7,8,9]
move_history = []
def box_to_xy(box):
    idx= box - 1 #Box 1: index 0
    row= 2 - (idx//3) #top row: 2, bottom: 0; 2 - so that box 1 is top left
    col= idx%3 #first col: 0
    x= col + 0.5
    y= row + 0.5
    return x,y
print("WELCOME TO TIC TAC TOE!")
#Display available positions
win_lines= [(1,2,3),(1,5,9),(1,4,7),(2,5,8),(3,6,9),(3,5,7),
               (7,8,9),(4,5,6)]    
#Check for X/O
def valid_character_check():
    char = 8
    while char not in ['X', 'O']:
        char = input("X or O?: ").strip().upper()
        if char not in ['X', 'O']:
            print("Incorrect character. Please enter X or O")
    return char
#Check for invalid cooridnates
def valid_position_check():
    position = (767, 767)
    while position not in available_boxes:
        #print(available_positions)
        user_input =  input("Please enter box # (1-9):")

        try:
            position = int(user_input.strip())
            #position = (float(x_str), float(y_str))

        except ValueError:
            print("Invalid entry. Please enter an integer from 1-9.")
            continue

        if position not in available_boxes:
            print("This position is not available. Please enter a different one.")
            #clear_output()

    return position

def get_player_history(move_history, char):

    return {pos for ch, pos in move_history if ch == char}
    
def updated_board(move_history, available_boxes):
    clear_output(wait = True)
    #Building Grid
    fig, ax = plt.subplots()
    ax.set_xlim(0, 3)
    ax.set_ylim(0, 3)
    ax.set_xticks([1, 2])
    ax.set_yticks([1, 2])
    ax.grid(True)
    ax.tick_params(axis='both', which='both', length=0) # Hide tick marks
    ax.set_xticklabels([]) # Hide x-axis labels
    ax.set_yticklabels([]) # Hide y-axis labels
    #Place available box numbers
    for box in available_boxes:
        x,y = box_to_xy(box)
        ax.text(x, y, str(box), ha='center', va='center', fontsize=18, color='gray', alpha = 0.35)
    #X/O Placement
    for ch, pos in move_history:
        x,y = box_to_xy(pos)
        ax.text(x, y, ch, ha='center', va='center', fontsize=30, color='black')
    plt.show()
    plt.close(fig)
    
def win_check(move_history, available_boxes):
    x_moves = get_player_history(move_history, 'X')
    o_moves = get_player_history(move_history, 'O')
    for line in win_lines:
        if set(line).issubset(x_moves):
            return 'X' 
        if set(line).issubset(o_moves):
            return 'O'
    return None
    
def keep_playing_choice():
    choice = 'invalid'
    while choice not in ['Y','N']:
        choice = input("Do you want to keep playing? (Y/N):").capitalize()
        #clear_output()
        if choice not in ['Y','N']:
            #clear_output()
            print("Invalid response. Please enter Y or N.")

    #clear_output()

    if choice == "Y":
        return True
    else:
        print("Game ended by user.")
        return False

updated_board(move_history, available_boxes)
#keep_playing
keep_playing = True

while keep_playing and available_boxes:
    #Have player enter letter and box
    char = valid_character_check()
    box = valid_position_check()
    move_history.append((char, box))
    available_boxes.remove(box)
    updated_board(move_history, available_boxes)
    winner = win_check(move_history, available_boxes)
    if winner:
        print(f"{winner} wins! Congratulations!")
        break
    if not available_boxes:
        print("Game has ended in a draw.")
        break
    keep_playing = keep_playing_choice()
```
